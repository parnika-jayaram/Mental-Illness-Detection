```python
YnewAX = YnewmodelData.drop(columns =[ "Anxiety","Stress","Depression"], axis=1)
YnewAy = YnewmodelData["Anxiety"]

YnewAX_train, YnewAX_test, YnewAY_train, YnewAY_test = train_test_split(YnewAX, YnewAy,
    test_size=0.2, random_state= 8)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
YnewAX_train=sc_X.fit_transform(YnewAX_train)
YnewAX_test=sc_X.transform(YnewAX_test)
```

```python
import keras
# Variance based feature importance
class VarImpVIANN(keras.callbacks.Callback):
    def __init__(self, verbose=0):
        self.verbose = verbose
        self.n = 0
        self.M2 = 0.0

    def on_train_begin(self, logs={}, verbose = 1):
        if self.verbose:
            print("VIANN version 1.0 (Wellford + Mean) update per epoch")
        self.diff = self.model.layers[0].get_weights()[0]

    def on_epoch_end(self, batch, logs={}):
        currentWeights = self.model.layers[0].get_weights()[0]

        self.n += 1
        delta = np.subtract(currentWeights, self.diff)
        self.diff += delta/self.n
        delta2 = np.subtract(currentWeights, self.diff)
        self.M2 += delta*delta2

        self.lastweights = self.model.layers[0].get_weights()[0]

    def on_train_end(self, batch, logs={}):
        if self.n < 2:
            self.s2 = float('nan')
        else:
            self.s2 = self.M2 / (self.n - 1)

        scores = np.sum(np.multiply(self.s2, np.abs(self.lastweights)), axis = 1)

        self.varScores = (scores - min(scores)) / (max(scores) - min(scores))
        if self.verbose:
            print("Most important variables: ",
                  np.array(self.varScores).argsort()[-10:][::-1])
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
VIANN = VarImpVIANN(verbose=1)
APYnewAmodel = Sequential()
APYnewAmodel.add(Dense(16, input_dim=42, activation='relu'))
# YnewAmodel.add(Dropout(0.5))
APYnewAmodel.add(Dense(12, activation='relu'))
# YnewAmodel.add(Dropout(0.5))
APYnewAmodel.add(Dense(8, activation='relu'))
# YnewAmodel.add(Dropout(0.5))
APYnewAmodel.add(Dense(1, activation='sigmoid'))
APYnewAmodel.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
APYnewAhistory = APYnewAmodel.fit(YnewAX_train, YnewAY_train, epochs=100, batch_size=64, validation_split=0.2, callbacks = [VIANN])
import matplotlib.pyplot as plt

# Plot the loss
plt.plot(APYnewAhistory.history['loss'], label='Training Loss')
plt.plot(APYnewAhistory.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
# plt.savefig('YnewALossVal_loss')

# Plot the accuracy
plt.plot(APYnewAhistory.history['accuracy'], label='Training Accuracy')
plt.plot(APYnewAhistory.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()
# plt.savefig('YnewAAccVal_acc')
```
